Use the pandas library to import the Data Frame.

```python
import pandas as pd
```

Load the Meesho_Payment_Sheet xlsx file after removing unnecessary data.

```python
df = pd.read_excel("/content/Meesho_Payment_Sheet_Oct1_Oct31(2).xlsx")
```

Obtaining a brief insight into the organization and contents of the DataFrame.

```python
df.head()
```

Creating a new Excel file named "Purchase_Price" and loading the data from the second sheet of the "Meesho_Payment_Sheet."

```python
df2 = pd.read_excel("/content/Purchase_Price.xlsx")
```

```python
df2.head()
```

Merge DataFrames: Merge the two DataFrames (df and df2) on the SKU column.

```python
merged_df = pd.merge(df, df2, left_on='SKU', right_on='Supplier SKU', how='inner')
```

```python
df.head()
```

**Calculate Revenue**

```python
merged_df['Total Revenue'] = (merged_df['Total Sale Amount (Incl. Commission & GST)'] -
                              merged_df['Sale Return Amount (Incl. GST)'] +
                              merged_df['Fixed Fee (Incl. GST)'] +
                              merged_df['Waas fee (inc Gst)'] +
                              merged_df['Shipping Revenue (Incl. GST)'] -
                              merged_df['Shipping Return Amount (Incl. GST)'] -
                              merged_df['Return premium (incl GST)'] -
                              merged_df['Return premium (incl GST) of Return'] +
                              merged_df['GST Compensation (PRP Shipping)'] +
                              merged_df['Shipping Charge (Excl. GST)'] +
                              merged_df['GST on Shipping Charge CGST + SGST on Shipping Charge'])
```

**Calculate Cost**

```python
merged_df['Total Cost'] = merged_df['Quantity'] * merged_df['Purchase Price']
```

**Calculate Profit**

```python
merged_df['Profit'] = merged_df['Total Revenue'] - merged_df['Total Cost']
```

**Group by SKU**

```python
sku_profitability = merged_df.groupby('SKU')['Profit'].sum().reset_index()
sku_profitability
```

**1. Total Profit for Entire Month**

```python
total_profit = merged_df['Profit'].sum()
print(f'Total Profit for the Entire Month: {total_profit}')
```

**2. Profitability SKU wise for Entire Month**

```python
print('Profitability SKU wise for Entire Month:')
print(sku_profitability)
```

**3. Total Payment Received for Entire Month:**

```python
total_payment_received = merged_df['Final Settlement Amount'].sum()
print(f'Total Payment Received for the Entire Month: {total_payment_received}')
```

2) Return Percentage

```python
return_percentage = (merged_df['Sale Return Amount (Incl. GST)'].sum() /
                    merged_df['Total Sale Amount (Incl. Commission & GST)'].sum()) * 100
print(f'Return Percentage: {return_percentage:.2f}%')
```

3) Average Order Value

```python
average_order_value = merged_df.groupby('Sub Order No')['Total Sale Amount (Incl. Commission & GST)'].mean().mean()
print(f'Average Order Value: {average_order_value:.2f}')
```

4) Sales Distribution by Product Category

```python
sales_distribution = merged_df.groupby('SKU')['Total Sale Amount (Incl. Commission & GST)'].sum()
print('Sales Distribution by Product Category:')
print(sales_distribution)
```
